```python
##Converting our generated JSON file to a pandas dataframe and cleaning it, changing days of the week from strings to integers

df = convert.read('medium.json')
df.loc[df.date == 'Tuesday', 'date'] = 2
df.loc[df.date == 'Monday', 'date'] = 1
df.loc[df.date == 'Wednesday', 'date'] = 3
df.loc[df.date == 'Thursday', 'date'] = 4
df.loc[df.date == 'Friday', 'date'] = 5
df.loc[df.date == 'Saturday', 'date'] = 6
df.loc[df.date == 'Sunday', 'date'] = 7
```

```python
## More data cleaning, changing times from strings to integers 

df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '12'), 'time'] = 0
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '01'), 'time'] = 1
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '02'), 'time'] = 2
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '03'), 'time'] = 3
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '04'), 'time'] = 4
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '05'), 'time'] = 5
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '06'), 'time'] = 6
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '07'), 'time'] = 7
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '08'), 'time'] = 8
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '09'), 'time'] = 9
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '10'), 'time'] = 10
df.loc[(df['time'].str[9:] == 'AM') & (df['time'].str[0:2] == '11'), 'time'] = 11
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '12'), 'time'] = 12
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '01'), 'time'] = 13
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '02'), 'time'] = 14
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '03'), 'time'] = 15
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '04'), 'time'] = 16
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '05'), 'time'] = 17
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '06'), 'time'] = 18
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '07'), 'time'] = 19
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '08'), 'time'] = 20
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '09'), 'time'] = 21
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '10'), 'time'] = 22
df.loc[(df['time'].str[9:] == 'PM') & (df['time'].str[0:2] == '11'), 'time'] = 23
```

```python
##Removing the entries with no coordinates
heat_df = df[df['points'].str.len() != 0]
```

```python
## Creating a list of lists of coordinates where each list represents the intersections for 1 day

heat_data_day = []
for i in range(1,8):
    heat_data1 = []
    for index, row in heat_df[heat_df['date'] == i].iterrows():
        for point in row['points']:
            lat_lon = [point[1], point[0]]
            heat_data1.append(lat_lon)
    heat_data_day.append(heat_data1)
```

```python
##Creating a list of lists of coordinates where each list represents the intersections for each hour 

heat_data_time = []
for i in range(0,24):
    heat_data2 = []
    for index, row in heat_df[heat_df['time'] == i].iterrows():
        for point in row['points']:
            lat_lon = [point[1], point[0]]
            heat_data2.append(lat_lon)
    heat_data_time.append(heat_data2)
```

```python
## Creating the map object using folium
map = folium.Map(location=[41.8781, -87.6298], zoom_start=11)
hm = folium.plugins.HeatMapWithTime(heat_data_time,auto_play=True,max_opacity=0.8)
hm.add_to(map)
```
